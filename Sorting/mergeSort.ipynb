```python
import numpy as np

def getarr(n):
    a = np.array(np.arange(0,n))
    np.random.shuffle(a)
    return a
def merge(A,p,q,r):
    i = p
    j = q+1
    B = np.copy(A[p:r+1])
    idx = 0
    while i<=q and j<=r:
        if A[i]<A[j]:
            B[idx] = A[i]
            idx = idx + 1
            i = i+1
        else:
            B[idx] = A[j]
            idx = idx + 1
            j = j+1
    if i>q:
        print('here')
        B[idx:] = A[j:r+1]
    elif j>r:
        print('here2')
        print(A,B,i,q,j,r,idx)        
        B[idx:] = A[i:q+1]
    print(B)
    A[p:r+1] = B
def mergeSort(A,p,r):
    # A[p...r] sorting
    if p<r:
        q = int((p+r)/2)
        mergeSort(A,p,q)
        mergeSort(A,q+1,r)
        merge(A,p,q,r)  # sorting 된 A[p..q], A[q+1..r]을 merge한다
    
```
