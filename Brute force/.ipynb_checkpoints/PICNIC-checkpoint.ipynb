```python
def checkpair(taken,pair):
    n = len(taken)
    firstFree = -1
    for i in range(0,n):
        if not taken[i]:
            firstFree = i
            break
    print('firstFree {}'.format(firstFree))
    if firstFree == -1:
        print('finished!')
        return 1
    ret = 0
    for pairWith in range(firstFree+1,n):
        if not taken[pairWith]:
            isfriend = False
            idx = [k for k,item in enumerate(pair) if item == pairWith]
            print('pairWith: {} , pair:{}'.format(pairWith,idx))
            for kk in idx:
                print('kk: {}'.format(kk))
                if (kk % 2 == 0 and (pair[kk+1] == firstFree)) or (kk % 2 == 1 and (pair[kk-1] == firstFree)):
                    
                    isfriend = True
                    taken[firstFree] = True
                    taken[pairWith] = True
                    print('lets recursive')
                    ret += checkpair(taken,pair)
                    print('out of recursive ret:{}'.format(ret))
                    isfriend = False
                    taken[firstFree] = False
                    taken[pairWith] = False
    print('finished all for')
    return ret
```
