```python
def CheckBoard(fw,Matrix,H,W,hstart,wstart,ridx):
    
    """
    모두 막혀져 있는 문제도 체크해야하나?    
    """
    for i in range(0,H):
        fw.writelines('{} \n'.format(Matrix[i][0:]))  
    fw.write('function start {}, {} \n'.format(hstart,wstart))
    fw.write('=='*20 + '\n')
    isdone = True
    for line in Matrix:
        if 1 in line:
            isdone = False
    if isdone:      
        print('finish')
        return 1
    for i in range(0,H):
        for j in range(0,W):
            if Matrix[i][j] == 1 and i <=hstart and j<wstart:
                fw.write('function exit {}, {} \n'.format(i,j))
                return 0
    ret = 0
    N = H*W   
    ridx = ridx + 1
    #print(ridx)
    for i in range(0,H):
        for j in range(0,W):            

            flag = False
            if i == hstart and j >= wstart:
                flag = True
            elif i > hstart:
                flag = True
            if flag:
                hidx = [i,i,i+1]
                widx = [j,j+1,j]                        
                htmp = [idx for idx,item in enumerate(hidx) if not (item>=0 and item <H)]
                wtmp = [idx for idx,item in enumerate(widx) if not (item>=0 and item <W)]            
                if not htmp and not wtmp:
                    if Matrix[hidx[0]][widx[0]] == 1 \
                    and Matrix[hidx[1]][widx[1]] == 1\
                    and Matrix[hidx[2]][widx[2]] == 1:
                        fw.write('{}'.format(hidx))
                        fw.write('{}'.format(widx))
                        fw.write('type1\n')
                        Matrix[hidx[0]][widx[0]] = 0
                        Matrix[hidx[1]][widx[1]] = 0
                        Matrix[hidx[2]][widx[2]] = 0
                        ret += CheckBoard(fw,Matrix,H,W,i,j,ridx)
                        Matrix[hidx[0]][widx[0]] = 1
                        Matrix[hidx[1]][widx[1]] = 1
                        Matrix[hidx[2]][widx[2]] = 1

                hidx = [i,i,i+1]
                widx = [j,j+1,j+1]                        
                htmp = [idx for idx,item in enumerate(hidx) if not (item>=0 and item <H)]
                wtmp = [idx for idx,item in enumerate(widx) if not (item>=0 and item <W)]            
                if not htmp and not wtmp:
                    if Matrix[hidx[0]][widx[0]] == 1 \
                    and Matrix[hidx[1]][widx[1]] == 1\
                    and Matrix[hidx[2]][widx[2]] == 1:
                        fw.write('{}'.format(hidx))
                        fw.write('{}'.format(widx))
                        fw.write('type2\n')
                        Matrix[hidx[0]][widx[0]] = 0
                        Matrix[hidx[1]][widx[1]] = 0
                        Matrix[hidx[2]][widx[2]] = 0
                        ret += CheckBoard(fw,Matrix,H,W,i,j,ridx)
                        Matrix[hidx[0]][widx[0]] = 1
                        Matrix[hidx[1]][widx[1]] = 1
                        Matrix[hidx[2]][widx[2]] = 1

                hidx = [i,i+1,i+1]
                widx = [j,j,j+1]                        
                htmp = [idx for idx,item in enumerate(hidx) if not (item>=0 and item <H)]
                wtmp = [idx for idx,item in enumerate(widx) if not (item>=0 and item <W)]            
                if not htmp and not wtmp:
                    if Matrix[hidx[0]][widx[0]] == 1 \
                    and Matrix[hidx[1]][widx[1]] == 1\
                    and Matrix[hidx[2]][widx[2]] == 1:
                        fw.write('{}'.format(hidx))
                        fw.write('{}'.format(hidx))
                        fw.write('{}'.format(widx))
                        fw.write('type3\n')
                        Matrix[hidx[0]][widx[0]] = 0
                        Matrix[hidx[1]][widx[1]] = 0
                        Matrix[hidx[2]][widx[2]] = 0
                        ret += CheckBoard(fw,Matrix,H,W,i,j,ridx)
                        Matrix[hidx[0]][widx[0]] = 1
                        Matrix[hidx[1]][widx[1]] = 1
                        Matrix[hidx[2]][widx[2]] = 1


                hidx = [i,i+1,i+1]
                widx = [j,j,j-1]                        
                htmp = [idx for idx,item in enumerate(hidx) if not (item>=0 and item <H)]
                wtmp = [idx for idx,item in enumerate(widx) if not (item>=0 and item <W)]            
                if not htmp and not wtmp:
                    if Matrix[hidx[0]][widx[0]] == 1 \
                    and Matrix[hidx[1]][widx[1]] == 1\
                    and Matrix[hidx[2]][widx[2]] == 1:
                        fw.write('{}'.format(hidx))
                        fw.write('{}'.format(widx))
                        fw.write('type4\n')
                        Matrix[hidx[0]][widx[0]] = 0
                        Matrix[hidx[1]][widx[1]] = 0
                        Matrix[hidx[2]][widx[2]] = 0
                        ret += CheckBoard(fw,Matrix,H,W,i,j,ridx)
                        Matrix[hidx[0]][widx[0]] = 1
                        Matrix[hidx[1]][widx[1]] = 1
                        Matrix[hidx[2]][widx[2]] = 1
      
    return ret
```

```python
f = open('BOARDCOVER.txt', 'r')
fw = open('BOARDCOVER_answer.txt', 'w')
idx = 0
n = int(f.readline())
while idx < n:
    tmp = f.readline().split()
    H = int(tmp[0])
    W = int(tmp[1])
    block = ''
    for i in range(H):
        tmp = f.readline().strip()                
        block = block + tmp 
    Matrix = [[0]*W for i in range(H)]
    for i in range(0,len(block)):
        ch = block[i]
        if ch == '.':
            widx = i%W
            hidx = int(i/W)
            Matrix[hidx][widx] = 1
        elif ch == '#':
            widx = i%W
            hidx = int(i/W)
            Matrix[hidx][widx] = 0
    for i in range(0,H):
        fw.writelines('{} \n'.format(Matrix[i][0:]))    
    fw.write('=='*20 + '\n')
    ret = CheckBoard(fw,Matrix,H,W,0,0,0)
    print(ret)
    idx = idx + 1    
fw.close()
f.close()
```
